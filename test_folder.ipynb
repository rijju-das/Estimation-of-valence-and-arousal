```python
import os
import shutil
import numpy as np
import pandas as pd
train_ann_file = 'train_annotation.csv'
val_dir = "val_set\\images"
train_dir = "train_set\\images"
train_img = []
for f in os.listdir(train_dir):
    train_img.append(f)
```

```python
for f in train_img:
    if(f == "340505.jpg"):
        print(f)
        anno = pd.read_csv(train_ann_file)
        row = anno[anno["filename"] == int(f.split(".")[0])]
        print(row)
```
